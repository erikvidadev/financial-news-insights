```python
import pandas as pd

stock_df = pd.read_csv("../data/processed/stock_data.csv", skiprows=[1])

cols_to_convert = ["Open", "Close", "High", "Low", "Volume"]
stock_df[cols_to_convert] = stock_df[cols_to_convert].apply(pd.to_numeric, errors="coerce")

 
stock_df
```

```python
stock_df.to_csv("../data/cleaned/cleaned_stock_data.csv", index=False)
```
